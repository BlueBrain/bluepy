# Usecase 2 - Plotting the location of selected cells

In this second use case we will look at how to show the location of the selected cells.

```python
import numpy as np
import matplotlib.pyplot as plt
from bluepy import Circuit, Cell


circuit_path = "/gpfs/bbp.cscs.ch/project/proj12/NSE/bluepy/circuits/CA1.O1/mooc-circuit/CircuitConfig"
circuit = Circuit(circuit_path)
```

Now we will query all neurons of one excitatory region and one inhibitory m-type, and plot their location projected onto the x-y plane.

```python
exc = circuit.cells.get({Cell.SYNAPSE_CLASS: "EXC", Cell.REGION: "mc5;SP"}, properties=[Cell.X, Cell.Y, Cell.Z])
inh = circuit.cells.get({Cell.SYNAPSE_CLASS: "INH", Cell.MTYPE: "SR_SCA"}, properties=[Cell.X, Cell.Y, Cell.Z])
```

```python
plt.plot(exc[Cell.X], exc[Cell.Y], '.r', label='exc')
plt.plot(inh[Cell.X], inh[Cell.Y], '.b', label='inh')

x_min = exc[Cell.X].max()
x_max = exc[Cell.X].min()

plt.ylabel('Layer')
plt.ylabel('y (um)')
plt.xlabel('x (um)')
plt.xlim([x_min, x_max])
plt.legend(numpoints=1);
```

And similarly projected onto the x-z plane:

```python
plt.gca().set_aspect('equal')

plt.plot(exc[Cell.X], exc[Cell.Z], '.r', label='exc')
plt.plot(inh[Cell.X], inh[Cell.Z], '.b', label='inh')

plt.xlabel('x (um)')
plt.ylabel('z (um)')
plt.legend(numpoints=1);
```

The hexagone shape for the red exc cells is due to the selection of an hexagonal sub-column of a O1 circuit.
